```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    "I've been waiting for a this my whole life.",
    "I hate this so much!",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
from transformers import AutoModel

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(checkpoint)
```

```python
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

### The AutoModel only has the transformer model without the head. The AutoModel only output vectors, the head of the model is needed to convert the vectors to logits

![Transformer_Model_API](https://huggingface.co/datasets/huggingface-course/documentation-images/resolve/main/en/chapter2/transformer_and_head-dark.svg)

```python
from transformers import AutoModelForSequenceClassification # Has built-in Sequence Classification head

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
```

```python
print(outputs.logits.shape)
print(outputs.logits)
```

```python
import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```
